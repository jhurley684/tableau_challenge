```python
import pandas as pd
```

```python
df = pd.read_csv (r'C:\Users\J\Desktop\tableau-challenge\CitiBikeData.csv')
print (df)
```

```python
# 834415 records too much for tableau and my machnine - reduce size by droping any NAN records

station_data = df.dropna()

station_data.head()
```

```python
# 834415 records too much for tableau and my machnine - reduce size by droping any \N in birthyear column
```

```python
# Reduce data set by using only records with birth year

has_birth_year = station_data[station_data["birth year"].str.contains("N")==False]
has_birth_year.info()
```

```python
# Getting rid of rows without birthyear is still more than 600k records - too much!
# Take a 30% randomized sample to use in Tableau. 

random_sample = has_birth_year.sample(frac = 0.3)

random_sample.info()
```

```python
# Export to use in Tableau (.csv)
random_sample.to_csv("random_bike_station_sample")
```
